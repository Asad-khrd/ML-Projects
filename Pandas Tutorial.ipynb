```python
df = pd.DataFrame(np.random.rand(300,5), index=np.arange(300)) 
```

```python
df
```

```python
df.describe()
```

```python
df.head(5)
```

```python
df.index # It will print indexes of the dataframe
```

```python
df.columns # It will print columns of the dataframe
```

```python
df.to_numpy() # It will create numpy array and stores the values of given dataframe into it
```

```python
df.T # It will transpose the dataframes (interchange columns and rows)
```

```python
df.sort_index(axis=0,ascending=False) # This will sort the rows in descending index order 
```

```python
df.sort_index(axis=1,ascending=False) # This will sort the columns in descending index order
```

```python
df[0] # This a series of column '0'
```

```python
type(df[0])
```

```python
df2 = df # Here,df2 act as view to df, it means that if we change any element of df2, the changes will also reflect in df
```

```python
df2[0][0] = 10
```

```python
df
```

```python
df2 = df.copy() # Here, df2 is independent of df, it means if we change any element of df2, the changes will not reflect in df
```

# Recommended Method to change the data in DataFrame

```python
df2.loc[0,0] = 1
```

```python
df2
```

# Deleting a column in dataframe

```python
df2 = df2.drop(4, axis=1) # Here, we have deleted column '4' of the dataframe
```
